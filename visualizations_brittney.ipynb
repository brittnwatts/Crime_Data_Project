```python
%matplotlib notebook

import pandas as pd
import hvplot.pandas
import matplotlib.pyplot as plt
import numpy as np
from bokeh.models import HoverTool
from holoviews import opts
import holoviews as hv
import warnings

warnings.filterwarnings("ignore")
```

```python
merged_df = pd.read_csv("Output/merged_data.csv")

merged_df.head()
```

```python
#finding the total crimes solved rate
total_actual = merged_df['actual'].sum()
total_cleared = merged_df['cleared'].sum()
solve_rate = total_cleared/total_actual

crimes_summary_df = pd.DataFrame({
    "Total Crimes Commited": [total_actual],
    "Total Crimes Soved": [total_cleared],
    "Solve Rate": [solve_rate]
    
})
crimes_summary_df
```

```python
#merged_columns = ["agency_name", "latitude", "longitude", "cleared", "actual"]
crime_df = merged_df.groupby("agency_name")[["latitude", "longitude", "cleared", "actual"]].first()
print(crime_df.head())
print(len(crime_df))
crime_df = crime_df.dropna()
crime_df = crime_df.fillna(value=0)
print(len(crime_df))
```

```python
map_plot = crime_df.hvplot.points(
    "longitude",
    "latitude",
    geo = True,
    title = "Crime Density Among Agencies in North Carolina",
    tiles = "OSM",
    frame_width = 700,
    frame_height = 500,
    size = "actual",
    scale = 0.90,
    color = "cleared",
    hover_cols=["agency_name"]
).opts(opts.Points(tools=['hover'], width=700, height=500))


map_plot
```
